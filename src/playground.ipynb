```python
# N = 0.5k - 5k, H = 1k - 40k, K = 5 - 10, F = 5 - 15, V = 1 - 200
def data_movement_lower_bound(N=5000, H=10000, K=7, F=10, V=100, D = 3):
    # lots of * 4, that's because vectors are encoded on 4 bytes (f32)

    # knn
    knn_bytes = H * K * D * 4

    # hyp/obs angles
    ang_bytes = H * K * D * 4 # could be fused so maybe this one is 0 theoretically

    # rotate displacements
    rotate_displ_bytes = H * D * D * 4

    # rotations matrices
    rot_mat_bytes = N * D * D * 4

    return knn_bytes + ang_bytes + rotate_displ_bytes + rot_mat_bytes
```
